```python
import pandas as pd

data = pd.read_csv('../Dataset/additional_features_iclr_2017_2018_2019_2020_with_CitationCount_EachYear.csv', encoding='latin-1')
data.dropna(inplace=True)
data
```

```python
def map_citation(row):
    year = row['year']
    return row[f'citedIn{year+1}']

data['citation'] = data.apply(map_citation, axis=1)
```

```python
data = data.drop(columns=['filename', 'year', 'title', 'citation_count', 'citedIn2018', 'citedIn2019', 'citedIn2020', 'citedIn2021', 'citedIn2022', 'citedIn2023', 'paperId', 'decision_ori', 'decision_final', 'rating1', 'rating2', 'rating3', 'mean', 'class_label', 'filetype2'], axis=1)
data = data[data.citation != 0]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(1, 10))
data = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
new_data = data[data['citation']>=7]
new_data
```

# Citing Sentence Predictor

```python
citing_sentence = new_data[['atr0', 'atr1', 'atr2', 'atr3', 'atr4', 'atr5', 'atr6', 'atr7', 'atr8', 'atr9', 'atr10', 'atr11', 'atr12', 'atr13', 'atr14', 'atr15', 'atr16', 'atr17', 'atr18', 'atr19']]
y = new_data['citation']
```

```python
from sklearn.feature_selection import f_regression, SelectKBest

def find_f(features, y, k):
    f_statistic = SelectKBest(score_func=f_regression, k=k)
    fit = f_statistic.fit(features, y)
    
    selected_feature_indices = f_statistic.get_support(indices=True)
    feature_names = features.columns
    chi2_scores = fit.scores_
    feature_score_dict = dict(zip(feature_names, chi2_scores))
    sorted_features = sorted(feature_score_dict.items(), key=lambda x: x[1], reverse=True)
    selected_feature_names = [feature[0] for feature in sorted_features[:k]]
    
    print("Selected Feature Names:", selected_feature_names)
    x_chi = f_statistic.fit_transform(features, y)
    
    x_train, x_test, y_train, y_test = train_split(x_chi, y)
    return x_train, x_test, y_train, y_test, selected_feature_names
```

```python
from sklearn.model_selection import train_test_split

def train_split(x, y):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
    return x_train, x_test, y_train, y_test
```

```python
test = pd.DataFrame(columns=['Features', 'MAE', 'RMSE'])
```

```python
from xgboost import XGBRegressor
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def model(features, y, i):
    model = XGBRegressor()
    x_train, x_test, y_train, y_test, selected_feature_names = find_f(features, y, i)
    model.fit(x_train, y_train)
    
    y_pred = model.predict(x_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    
    new_data = pd.DataFrame({'Features': [selected_feature_names], 'MAE': [mae], 'RMSE': [rmse]})
    
    global test
    test = pd.concat([test, new_data], ignore_index=True)
    
    print(f'Mean Absolute Error (MAE): {mae:3f}')
    print(f'Root Mean Squared Error (RMSE): {rmse:3f}')
```

```python
for i in dict(enumerate(citing_sentence)):
    model(citing_sentence, y, i+1)
```

# Regular Sentences Predictor

```python
regular_sentences = new_data[['atr20', 'atr21', 'atr22', 'atr23', 'atr24', 'atr25', 'atr26', 'atr27', 'atr28', 'atr29', 'atr30', 'atr31', 'atr32', 'atr33', 'atr34', 'atr35', 'atr36', 'atr37', 'atr38', 'atr39']]
```

```python
for i in dict(enumerate(regular_sentences)):
    model(regular_sentences, y, i+1)
```

# Reference-Based Predictor

```python
reference_based = new_data[['num_ref', 'num_ref_3years', 'arxiv', 'neurips', 'iclr', 'icml', 'aaai', 'iccv', 'cvpr', 'emnlp', 'acl', 'naacl', 'eccv', 'icra', 'icassp', 'ijcai', 'aistats', 'sigkdd', 'neuralcom', 'ieee_tran', 'acm_tran', 'mit_press', 'nature', 'jmlr']]
```

```python
for i in dict(enumerate(reference_based)):
    model(reference_based, y, i+1)
```

# Combination Predictor

```python
combination = citing_sentence.join([regular_sentences, reference_based])
```

```python
for i in dict(enumerate(combination)):
    model(combination, y, i+1)
```
